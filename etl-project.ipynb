```python
import pandas as pd
from sqlalchemy import create_engine 
```

## Extract Data

```python
# A comprehensive list of movies available on various streaming platforms (IMDb dataset)
streaming_csv = "Resources/MoviesOnStreamingPlatforms_updated.csv"
streaming_df_orig = pd.read_csv(streaming_csv)
streaming_df_orig.head()
```

```python
# The CSV file contains 85,855 movies with 22 attributes
movies_csv = "Resources/imdb_movies.csv"
movies_df_orig = pd.read_csv(movies_csv, low_memory=False)
movies_df_orig.head()
```

```python
# The CSV file contains 297,705 cast members with 20 attributes
names_csv = "Resources/imdb_names.csv"
names_df_orig = pd.read_csv(names_csv)
names_df_orig.head()
```

```python
# This CSV file contains 835,513 cast members roles in movies with 6 attributes
titles_csv = "Resources/imdb_title_principals.csv"
titles_df_orig = pd.read_csv(titles_csv)
titles_df_orig.head()
```

## Clean Data

#### Streaming DataFrame

```python
# Format Streaming data column names
streaming_df_1 = streaming_df_orig.copy()
streaming_df_1.columns = streaming_df_1.columns.str.lower()
streaming_df_1.head()
```

```python
# Clean Streaming data column names
streaming_df_2 = streaming_df_1.rename(columns={
    'age': 'target_age',
    'imdb': 'imdb_rating',
    'rotten tomatoes': 'rotten_tomatoes_converted',
    'prime video': 'prime_video'
})

streaming_df_3 = streaming_df_2.drop(columns='unnamed: 0')
streaming_df_3.head()
# streaming_df_3.dtypes
```

```python
# Clean 'rotten_tomatoes' rating to comparable unit as 'imdb' rating to allow aggregation and comparison
# Rotten Tomatoes rating was initially percentage (max 100%) while IMDb Metascore rating is absolute value (max 10).

streaming_df_4 = streaming_df_3.copy()
streaming_df_4['rotten_tomatoes_converted'] = streaming_df_4['rotten_tomatoes_converted'].str.replace(r'%', r'.0').astype('float') / 10.0
# Divide by 10 instead of 100 to convert percentage to absolue value with max rating of 10

streaming_df_4.head()
# streaming_df_4.dtypes
```

#### Movies DataFrame

```python
# Clean column data since many datapoints overlap with Streaming DataFrame
movies_df_1 = movies_df_orig[['imdb_title_id', 'title', 'year', 'description']].copy()
movies_df_1.head()
# movies_df_1.dtypes
```

```python
# Locate non-integer year from Movies df preventing series reclass to integer necessary to merge
movies_df_1.loc[movies_df_1['year'] == 'TV Movie 2019']
```

```python
# Clean year value preventing merge by replacing with year value
movies_df_2 = movies_df_1.replace(to_replace='TV Movie 2019', value='2019')
movies_df_2.loc[movies_df_2['title'] == 'Bad Education']
```

```python
# Reclass year value to integer
movies_df_2['year'] = movies_df_2['year'].astype('int64')
movies_df_2.dtypes
```

```python
# Inner merge to only view movies available on streaming platforms
streaming_df_merged = streaming_df_4.merge(movies_df_2, how='inner', on=['title','year'])
streaming_df_merged.set_index('imdb_title_id', inplace=True)
streaming_df_clean = streaming_df_merged.drop('id', 1)
streaming_df_clean
```

#### Names DataFrame

```python
names_df_1 = names_df_orig.copy()

# Split spouses_string to find most recent spouse
print(f"spouses max value: {names_df_1['spouses'].max()}")
names_df_1.head()
```

```python
# Split out most recent spouse name
names_df_2 = names_df_1.copy()
names_df_2['spouse_last'] = names_df_2['spouses_string'].str.split('(', expand=True)[0]
names_df_2.head()
```

```python
names_df_3 = names_df_2.copy()
names_df_3['place_of_birth_1'] = names_df_2['place_of_birth'].str.split(',', expand=True)[0]
names_df_3['place_of_birth_2'] = names_df_3['place_of_birth'].str.split(',', expand=True)[1]
names_df_3['place_of_birth_3'] = names_df_3['place_of_birth'].str.split(',', expand=True)[2]
names_df_3['place_of_birth_4'] = names_df_3['place_of_birth'].str.split(',', expand=True)[3]
names_df_3.head()
```

```python
names_df_4 = names_df_3.copy()
names_df_4['place_of_death_1'] = names_df_3['place_of_death'].str.split(',', expand=True)[0]
names_df_4['place_of_death_2'] = names_df_3['place_of_death'].str.split(',', expand=True)[1]
names_df_4['place_of_death_3'] = names_df_3['place_of_death'].str.split(',', expand=True)[2]
names_df_4['place_of_death_4'] = names_df_3['place_of_death'].str.split(',', expand=True)[3]
names_df_4.head()
```

```python
# Rename to clarify column unit of measure
names_df_5 = names_df_4.rename(columns={
    'height': 'height_cm'
})

# Drop unnecessary columns and reorder
names_df_5 = names_df_5[['imdb_name_id',
                         'name',
                         'birth_name',
                         'height_cm',
                         'bio',
                         'date_of_birth',
                         'place_of_birth_1',
                         'place_of_birth_2',
                         'place_of_birth_3',
                         'place_of_birth_4',
                         'date_of_death',
                         'place_of_death_1',
                         'place_of_death_2',
                         'place_of_death_3',
                         'place_of_death_4',
                         'reason_of_death',
                         'spouse_last',
                         'spouses',
                         'divorces',
                         'spouses_with_children',
                         'children'
                        ]]

names_df_5.set_index('imdb_name_id', inplace=True)
names_df_5
```

#### Titles DataFrame

```python
titles_df_1 = titles_df_orig.copy()
titles_df_1.set_index('imdb_title_id', inplace=True)
titles_df_1
```
